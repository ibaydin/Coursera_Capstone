```python
import pandas as pd 
import numpy as np 
from geopy.geocoders import Nominatim
```

```python
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)[0]
```

    Next cell implements the following requirements:    
    * Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.    
    * If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df=df[df['Borough']!='Not assigned'].reset_index(drop=True)
df.loc[df['Neighbourhood'] == "Not assigned", "Neighbourhood"] = df['Borough']
```

Let's add Latitude and Longitude columns

```python
df['Latitude']=0
df['Longitude']=0
```

```python
df.head()
```

Let's get "Latitude and Longitude " for the boroughs

```python
def get_location(brgh):
    geolocator = Nominatim(user_agent="explorer")
    location = geolocator.geocode('{}, Toronto, Ontario'.format(brgh) )
    latitude = location.latitude
    longitude = location.longitude
    return latitude,longitude
```

```python
for i in df.index:
    r=get_location(df['Borough'][i])
    df.loc[i,'Latitude'] = r[0]
    df.loc[i,'Longitude'] = r[1]
    #print(df['Borough'][i])

print('done')
        
```
